```python
import warnings
import MDAnalysis
import numpy as np
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore', category=DeprecationWarning, message='.*use @default decorator instead.*')
%matplotlib inline
#c = get_config()
# ... Any other configurables you want to set
#c.InteractiveShellApp.matplotlib = "inline"
```

```python
# Loading the molecular data:
#traj = MDAnalysis.coordinates.DCD.DCDReader("../1e/trajectory.dcd")
u = MDAnalysis.Universe("1e_lastframe.pdb","../1e/trajectory.dcd")
```

```python
# Testing how MDAnalysis works:
lig = u.select_atoms("resname 1ED")
```

```python
# Computing the radius of gyration:
Rgyr = []
atms = u.select_atoms("all")
for ts in u.trajectory:
    Rgyr.append((u.trajectory.time,atms.radius_of_gyration()))
Rgyr = np.array(Rgyr)    
```

```python
plt.plot(Rgyr[:,0], Rgyr[:,1], 'r--', lw=2, label=r"$R_G$")
plt.xlabel("time (ps)")
plt.ylabel(r"radius of gyration $R_G$ ($\AA$)")
plt.savefig("Rgyr.pdf")
plt.show()
```
